```python
import pandas as pd

filename = r'C:\Users\Lee\Desktop\Project\천안시 시각화\천안시_DataDir\천안시apt.xlsx'

house_loc = pd.read_excel(filename,engine='openpyxl',skiprows=[1,2],header=1)
house_loc =house_loc.drop(['사업승인일','사용검사일'],axis=1)
house_loc
```

```python
house_loc['층수'].fillna(6,inplace=True)
house_loc['층수']=house_loc['층수'].astype(int)
```

```python
sorted_house_loc = house_loc.sort_values(by=['세대수'],ascending=[False])
sorted_house_loc
```

```python
sorted_house_loc[sorted_house_loc['공동주택명']=='두정역이안더센트럴아파트']
```

```python
# 세대수를 기준으로 1분위수에 해당하는 값만 뽑기
q1 = house_loc['세대수'].quantile([0.25, 0.5, 0.75, 1.0])
# q1
selected_values = house_loc[house_loc['세대수'] >=q1[0.50]]
q_house_loc = selected_values.sort_values(by=['세대수'],ascending=False)
q_house_loc['도로명주소'] = q_house_loc['도로명주소'].str[:-5]
q_house_loc
```

```python
from geopy.geocoders import Nominatim
```

```python
geolocator =  Nominatim(user_agent='South Korea',timeout=10)
def get_coordinates(address):
    location = geolocator.geocode(address)
    if location:
        return location.latitude, location.longitude
    else:
        return None, None
q_house_loc['위도'], q_house_loc['경도'] = zip(*q_house_loc['도로명주소'].apply(get_coordinates))
q_house_loc
```

```python
from itertools import product

distance_df = pd.DataFrame(product(bus['정류장명'],q_house_loc['공동주택명']),columns=['정류장명','공동주택명'])
m1_df = distance_df.merge(bus[['정류장명','위도','경도']],on='정류장명',how='left')
m1_df.rename(columns={'위도':'버스정류장_위도','경도':'버스정류장_경도'},inplace=True)

# distance_df
m1_df
```

```python
m2_df = m1_df.merge(q_house_loc[['공동주택명','위도','경도']],on='공동주택명',how='left')
m2_df.rename(columns={'위도':'공동주택명_위도','경도':'공동주택명_경도'},inplace=True)
m2_df
```

```python
import geopy.distance
m3_df = m2_df.copy()
m3_df['버스'] = pd.Series(zip(m3_df['버스정류장_위도'],m3_df['버스정류장_경도']))
m3_df['주택위치'] = pd.Series(zip(m3_df['공동주택명_위도'],m3_df['공동주택명_경도']))
m3_df['distance'] = m3_df.apply(lambda elm : geopy.distance.distance(elm['버스'],elm['주택위치']).km, axis=1) # .km -> km로 설정
bus_house_distance = m3_df.copy()
```
