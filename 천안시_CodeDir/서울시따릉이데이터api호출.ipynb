# 좌표 -> 반경 2km 내의 카테고리 갯수

```python
CateGroupCode = ['MT1','CS2','PS3','SC4','AC5','PK6','OL7','SW8','BK9','CT1','AG2','PO3','AT4','AD5','FD6','CE7','HP8','PM9']
code_df = pd.DataFrame(columns=CateGroupCode)
bike_station_location_cp = pd.concat([bike_station_location_cp,code_df],axis=1)
bike_station_location_cp
```

```python

apiKey = '1d49e941371e685a965fba06520cda68'
cnt=0
# 적용할 데이터프레임
# bus_code_df = 데이터프레임.copy()

run_df = bike_station_location_cp.copy()

# run_df = run_df.head()

headers = {'Authorization': 'KakaoAK {}'.format(apiKey) }
for code in tqdm(CateGroupCode):
  for idx in tqdm(range(len(run_df))):
    x = run_df.at[idx,'경도']
    y = run_df.at[idx,'위도']
    try:
      searching ='category_group_code={code}&x={x}&y={y}&radius={radius}'.format(code=code,x=str(x),y=str(y),radius=1000)
      url='https://dapi.kakao.com/v2/local/search/category.json?{}'.format(searching)
      data=requests.post(url,headers=headers)
      result = json.loads(data.text)
      # print(result)    

      run_df.at[idx,code] = result.get('meta').get('total_count')


    except:
      cnt+1
      if cnt%50 ==0:
        print('안긁히고 있는 데이터 갯수:',cnt)
      pass
  print(code,'코드 완료')
```
